## Distribution Plot

- Rugplot
- Histogram
- KDE Plot

```python
df.head()
```

```python
plt.figure(figsize=(10,7), dpi=300)

sns.rugplot(x="salary",data=df, height=.5)
```

```python
sns.set(style='ticks')
sns.displot(data=df,x="salary", bins=20,
           color="blue",edgecolor="black",
           lw=4,ls='--', kde=True,rug=True)
plt.show()
```

```python
sns.histplot(data=df,x='salary')
```

```python
sns.kdeplot(data=df,x="salary")
```

## Countplot & Barplot

- Countplot() -- Counts Number of rows per category
- barplot() -- General form of displaying any chosen metric per category

```python
df.head()
```

```python
df["division"].value_counts()
```

```python
plt.figure(figsize=(10,7),dpi=200)
sns.countplot(data=df,x="level of education", edgecolor="black",
             hue="division", palette="Set2")
plt.legend(loc='right')
#plt.ylim(90,260)
```

```python
df["level of education"].value_counts()
```

```python
plt.figure(figsize=(10,7),dpi=200)
sns.barplot(data=df,x="level of education",y="salary",
            estimator=np.mean,ci='sd',edgecolor="black",
           hue="division")
plt.legend(bbox_to_anchor=(1.05,1));
```

### Boxplot
- Display the distribution of contineous variable 
- Does through use of quartiles
- Quartiles separate out data into 4 equal number of data point

    -- 25% of datapoints in bottom quartile
    
    -- 50th percentile in median
    
- Boxplot quickly gives statistical distribution information in visual format

```python
 df=pd.read_csv("C:\\Users\\surani\\Machine_Learning_Python\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv")
```

```python
df
```

```python
plt.figure(figsize=(10,7),dpi=300)
sns.boxplot(data=df,y="reading score",x="parental level of education",
           hue="test preparation course",
           palette="Set2")
plt.legend(bbox_to_anchor=(1.05,0.5))
plt.show()
```

```python
plt.figure(figsize=(10,7),dpi=300)
sns.violinplot(data=df,x="reading score",y="parental level of education",
           hue="test preparation course",
           palette="Set2")
plt.legend(bbox_to_anchor=(1.05,0.5))
plt.show()
```

```python
plt.figure(figsize=(10,7),dpi=300)
sns.violinplot(data=df,x="reading score",y="parental level of education",
           hue="test preparation course",split=True,
           palette="Set2")
plt.legend(bbox_to_anchor=(1.05,0.5))
plt.show()
```
